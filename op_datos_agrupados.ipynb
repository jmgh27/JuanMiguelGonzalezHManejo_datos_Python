```python
ratings = pd.read_csv(r'C:\Users\jmgh8\py_analisis_de_datos\2daparte\032_ratings.csv')

peliculas = pd.read_csv(r'C:\Users\jmgh8\py_analisis_de_datos\2daparte\032_peliculas.csv')

usuarios = pd.read_csv(r'C:\Users\jmgh8\py_analisis_de_datos\2daparte\032_usuarios.csv')
```

```python
#Juan Miguel Gonzalez Hernandez
peli_ratings = pd.merge(peliculas, ratings)
clasif = pd.merge(peli_ratings, usuarios)
```

```python
mas_ratings = clasif.groupby('titulo').size().sort_values(ascending=False)[:25]
mas_ratings
```

```python
#Juan Miguel Gonzalez Hernandez
peli_stats = clasif.groupby('titulo').agg({'rating': [np.size,np.mean]})
peli_stats.head()
```

```python
peli_stats.sort_values([('rating','mean')], ascending=False).head()
```

```python
#Juan Miguel Gonzalez Hernandez
minimo_100 = peli_stats['rating']['size']>= 100
```
